**1. Import Library**

```python
import matplotlib.pyplot as plt  # library untuk membuat visualisasi (grafik)
```

**2. Buat Dataset**

```python
suhu = [20, 22, 24, 26, 28, 30, 32, 34, 36]        # data suhu sebagai sumbu X
penjualan = [50, 60, 70, 80, 90, 100, 110, 120, 130]  # data penjualan sebagai sumbu Y
```

**3. Basic Scatter Plot**

```python
plt.scatter(suhu, penjualan)   # membuat scatter plot dari dua variabel
plt.title('Hubungan Penjualan Es Krim dengan Suhu')  # judul grafik
plt.xlabel('Suhu')             # label sumbu X
plt.ylabel('Penjualan')        # label sumbu Y
plt.show()                     # tampilkan grafik
```

**4. Kustomisasi Scatter Plot**

```python
plt.scatter(
    suhu,
    penjualan,
    color='orange',     # warna titik
    s=100,              # ukuran titik
    edgecolor='black',  # warna pinggiran titik
    alpha=0.7           # transparansi titik
)
plt.title('Hubungan Penjualan Es Krim dengan Suhu')  # judul grafik
plt.xlabel('Suhu')                                   # label sumbu X
plt.ylabel('Penjualan')                              # label sumbu Y
plt.grid(True)                                       # menampilkan grid
plt.show()                                           # tampilkan grafik
```

**5. Multiple Scatter Plot**

```python
suhu = [20, 22, 24, 26, 28, 30, 32, 34, 36]  # data suhu

# dataset penjualan berdasarkan jenis hari
penjualan_weekdays = [50, 60, 70, 80, 90, 100, 110, 120, 130]  # hari kerja
penjualan_weekends = [40, 50, 60, 70, 80, 90, 100, 110, 120]   # akhir pekan

# plot scatter hari kerja
plt.scatter(
    suhu,
    penjualan_weekdays,
    color='green',
    label='Hari Kerja',
    s=80                 # ukuran titik
)

# plot scatter akhir pekan
plt.scatter(
    suhu,
    penjualan_weekends,
    color='purple',
    label='Akhir Pekan',
    s=80
)

plt.title('Hubungan Penjualan Es Krim dengan Suhu')  # judul
plt.xlabel('Suhu')                                   # label X
plt.ylabel('Penjualan')                              # label Y
plt.legend()                                         # tampilkan legenda
plt.show()                                           # tampilkan plot
```

**6. Melihat Hubungan Data dengan Beberapa Variabel**

```python
suhu = [20, 22, 24, 26, 28, 30, 32, 34, 36]          # data suhu
penjualan = [50, 60, 70, 80, 90, 100, 110, 120, 130]  # data penjualan

# dataset variabel tambahan
kelembapan = [60, 65, 70, 75, 80, 85, 90, 95, 100]    # warna berdasarkan kelembapan
ukuran_titik = [100, 120, 140, 160, 180, 200, 220, 240, 260]  # ukuran titik bervariasi

# scatter plot dengan 3 variabel (X, Y, warna, ukuran)
plt.scatter(
    suhu,
    penjualan,
    c=kelembapan,          # warna berdasarkan kelembapan
    s=ukuran_titik,        # ukuran titik
    cmap='coolwarm',       # colormap
    alpha=0.6              # transparansi
)

plt.title('Hubungan Penjualan Es Krim dengan Suhu')  # judul
plt.xlabel('Suhu')                                   # label X
plt.ylabel('Penjualan')                              # label Y
plt.colorbar(label='Kelembapan (%)')                 # legenda warna
plt.show()                                           # tampilkan grafik
```

**7. 3D Scatter Plot**

```python
import matplotlib.pyplot as plt                   # library plot
from mpl_toolkits.mplot3d import Axes3D           # modul untuk grafik 3D

suhu = [20, 22, 24, 26, 28, 30, 32, 34, 36]        # sumbu X

# dataset penjualan
penjualan_weekdays = [50, 60, 70, 80, 90, 100, 110, 120, 130]   # sumbu Y
penjualan_weekends = [40, 50, 60, 70, 80, 90, 100, 110, 120]    # sumbu Z

# membuat figure 3D
fig = plt.figure(figsize=(15, 13))               # ukuran figure
ax = fig.add_subplot(111, projection='3d')        # area plot 3D

# scatter plot 3D
ax.scatter(
    suhu,
    penjualan_weekdays,
    penjualan_weekends,
    color='green',
    label='Data Penjualan',
    s=80
)

ax.set_title('Hubungan Penjualan Es Krim dengan Suhu (3D Ver)')  # judul
ax.set_xlabel('Suhu')                                             # label X
ax.set_ylabel('Penjualan Weekdays')                               # label Y
ax.set_zlabel('Penjualan Weekends')                               # label Z

plt.legend()   # tampilkan legenda
plt.show()     # tampilkan plot
```
